```python
data_folder = "https://github.com/PLombardo811/baseballdatabank/blob/master/upstream/"
Team_url = data_folder + "Teams.csv?raw=true"
Team_content = requests.get(Team_url).content
Team_csv = pd.read_csv(io.StringIO(Team_content.decode('utf-8')))

print(Team_csv.head())
```

```python
Team_csv = Team_csv[Team_csv['lgID'].notnull()]
Team_csv["yearID"] = Team_csv["yearID"].apply(str)

print(Team_csv.head())
```

```python
Team_pairs = Team_csv[["yearID", "franchID"]].copy()
Team_pairs["edges"] = Team_pairs.apply(lambda x: (x["franchID"], x["yearID"]), axis=1)

G = nx.DiGraph()
G.add_edges_from(Team_pairs["edges"].tolist(), color = 'g')
G.add_nodes_from(Team_csv["franchID"].tolist()+Team_csv["yearID"].tolist(), color='y')

pos=nx.spring_layout(G, iterations=50)

offset=0.05
pos_labels = {}
keys = pos.keys()
for key in keys:
    x, y = pos[key]
    pos_labels[key] = (x, y+offset)

plt.figure(figsize=(17,12))
d = nx.degree(G)
plt.hist(d.values(),bins=len(pos))
plt.xticks(range(1,len(pos)))
plt.show()
```

```python
c = nx.average_shortest_path_length(G)
print("Characteristic path length of 10x10 mesh = {ave} \n").format(ave = c)
```

```python
plt.figure(figsize=(19,13))
pos = nx.spring_layout(G,k=.12)
default_distance_between_nodes = 1.0/(math.sqrt(len(G.nodes())))
nx.draw_networkx_edges(G,pos, edge_color='b',alpha=.3)  
nx.draw_networkx_nodes(G,pos,node_color='c',node_size=300)
nx.draw_networkx_nodes(G,pos,pos_labels,node_color='r',node_size=300)
nx.draw_networkx_labels(G,pos,pos_labels,font_size=10,font_color='k')
plt.axis('off')
plt.show()
```
